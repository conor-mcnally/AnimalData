```python
import pandas as pd
import os
import folium
```

```python
df = pd.read_json('animals.json')

lat = [21, 34, -6, 21, -3, 29, -3, -14, -18, -22, 27, 77]
lon = [90, 103, -20, 92, 23, 83, 112, -157, 23, 14, 87, -94]

df['lat'] = lat
df['lon'] = lon
df['lat'] = df['lat'].astype(float)
df['lon'] = df['lon'].astype(float)
```

```python
animals = df.filter(['name', 'location', 'lat', 'lon', 'population', 'status', 'image'])
```

```python
url = animals['image'].apply(lambda x: x['url'])
animals['image'] = url
animals
```

```python
#Create blank map
m = folium.Map(tiles='Stamen Watercolor')

#Markers

for (index, row) in animals.iterrows():
    
    icon = folium.features.CustomIcon(f"http://www.bloowatch.org{row.loc['image']}", icon_size=(30,30))
    
    folium.Marker(
        location = [row.loc['lat'], row.loc['lon']],
        icon = icon,
        popup = "Population:\n " + row.loc['population'] + "\nStatus:\n " + row.loc['status'],
        tooltip = row.loc['name']
    ).add_to(m)

#Print map
m
```
